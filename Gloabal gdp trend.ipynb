```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Imputation Method
imputer = SimpleImputer(strategy='mean')

# Applying the imputer to the DataFrame
# Excluding the 'Country' column as it's categorical and has no missing values
GDP = data.copy()
GDP.iloc[:, 1:] = imputer.fit_transform(GDP.iloc[:, 1:])

# Now, GDP is your new DataFrame with missing values imputed
GDP.head()
```

```python
# Aggregating GDP by year
years = [str(year) for year in range(1980, 2029)]
global_gdp = data[years].sum()

# Creating a DataFrame for the plot
gdp_trends = pd.DataFrame({
    'Year': years,
    'Global GDP': global_gdp
})
print(gdp_trends.head())
print(gdp_trends.tail())
```

```python
import pycountry

# Function to get ISO code for a given country
def get_iso_code(country_name):
    try:
        country = pycountry.countries.get(name=country_name)
        return country.alpha_3
    except AttributeError:
        return None

# Add ISO code column to the DataFrame
data['ISO_Code'] = data['Country'].apply(get_iso_code)

# Display the updated DataFrame
print(data.head())
```

```python
# Filtering out the countries of interest
countries_of_interest = ['Denmark', 'Iceland ','India','Ireland','Luxembourg','Norway','Qatar','Singapore','Sweden','Switzerland','United States']
filtered_data = data[data['Country'].isin(countries_of_interest)]
print(filtered_data.head())
# Creating a DataFrame with necessary data
map_data = pd.DataFrame({
    'Country': filtered_data['Country'],
    'GDP': filtered_data['2028'],
    'ISO Alpha':filtered_data['ISO_Code']  # ISO Alpha-3 country codes
})

# Creating the interactive choropleth map
fig = px.choropleth(
    map_data,
    locations='ISO Alpha',
    color='GDP',
    hover_name='Country',
    color_continuous_scale=px.colors.sequential.Plasma,
    title='GDP by Country for the Year 2028'
)

# Customizing the layout
fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    ),
    annotations=[dict(
        x=0.55,
        y=0.1,
        xref='paper',
        yref='paper',
        text='Source: <a href="https://www.imf.org/">\
            International Monetary Fund</a>',
        showarrow=False
    )]
)

# Show the plot
fig.show()
```
